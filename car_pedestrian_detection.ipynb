```python
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# TRAINING
model = YOLO('yolov8m.pt')
model.train(
    data='C:/Users/Sourabh/Downloads/main data/data.yaml',
    epochs=6,
    patience=5,
    batch=8,
    imgsz=450,
    device='cpu',
    workers=4
)
```

```python
# VALIDATION
# === VALIDATION ===
model = YOLO("runs/detect/train/weights/best.pt")
metrics = model.val(
    data='C:/Users/Sourabh/Downloads/main data/data.yaml',
    imgsz=450,
    batch=8,
    device="cpu",
    workers=4
)
```

```python
print(f"mAP@.5:.95 : {metrics.box.map:.4f}")
print(f"mAP@.5     : {metrics.box.map50:.4f}")
print(f"mAP@.75    : {metrics.box.map75:.4f}")
```

```python
# === PREDICTION ON TEST SET ===
results = model.predict(
    source='C:/Users/Sourabh/Downloads/main data/test/images',
    conf=0.5,
    iou=0.75,
    device="cpu"
)
```

```python
# === SAVE PREDICTIONS ===
prediction_dir = "C:/Users/Sourabh/Downloads/main data/predictions"
os.makedirs(prediction_dir, exist_ok=True)
```

```python
test_img_list = []

for result in results:
    if len(result.boxes.xyxy):
        name = os.path.basename(result.path).split(".")[0]
        boxes = result.boxes.xyxy.cpu().numpy()
        scores = result.boxes.conf.cpu().numpy()

        test_img_list.append(name)

        label_file_path = os.path.join(prediction_dir, name + ".txt")
        with open(label_file_path, "w") as f:
            for score, box in zip(scores, boxes):
                text = f"{score:.4f} " + " ".join(map(lambda x: f"{x:.2f}", box))
                f.write(text + "\n")
```

```python
# === VISUALIZE PREDICTIONS ===
def show_bbox(img, boxes, scores, axis, color=(0, 255, 0)):
    boxes = boxes.astype(int)
    img = img.copy()
    for i, box in enumerate(boxes):
        score = f"{scores[i]:.2f}"
        cv2.rectangle(img, (box[0], box[1]), (box[2], box[3]), color, 2)
        y = box[1] - 10 if box[1] - 10 > 10 else box[1] + 10
        cv2.putText(img, score, (box[0], y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    
    axis.imshow(img)
    axis.axis("off")
```

```python
def show_bbox(img, boxes, scores, axis):
    for box, score in zip(boxes, scores):
        x1, y1, x2, y2 = box
        axis.add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                     edgecolor='lime', facecolor='none', linewidth=2))
        axis.text(x1, y1, f"{score:.2f}", color='lime', fontsize=8)
    axis.imshow(img)
    axis.axis('off')

# Paths
image_dir = "C:/Users/Sourabh/Downloads/main data/test/images"
label_dir = "C:/Users/Sourabh/Downloads/main data/predictions"

# Image name prefixes
test_img_list = [
    "2024-08-05-173000_2024-08-05-174459_1000027-1-0-9_mp4-0060_jpg",
    "2024-08-05-173000_2024-08-05-174459_1000027-1-0-9_mp4-0072_jpg",
    # Add more if needed
]

# Setup plot
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
plt.subplots_adjust(wspace=0.1, hspace=0.1)
ax = axes.flatten()

# Separate subplot index
plot_index = 0

# Process each image prefix
for prefix in test_img_list:
    matched_file = next((f for f in os.listdir(image_dir) if f.startswith(prefix)), None)

    if matched_file is None:
        print(f"Image starting with {prefix} not found.")
        continue

    img_path = os.path.join(image_dir, matched_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    label_file_path = os.path.join(label_dir, prefix + ".txt")
    if not os.path.exists(label_file_path):
        print(f"Label file missing for: {prefix}")
        continue

    label = pd.read_csv(label_file_path, sep=" ", header=None).values
    scores = label[:, 0]
    boxes = label[:, 1:]

    if plot_index < len(ax):
        show_bbox(img, boxes, scores, axis=ax[plot_index])
        plot_index += 1

# Hide unused axes
for i in range(plot_index, len(ax)):
    ax[i].axis('off')
plt.tight_layout()
plt.savefig("car.png", bbox_inches='tight', pad_inches=0)
plt.show()
```
